```python
import numpy as np
import pandas as pd
import random
import string
```

```python
df = pd.read_csv('../data/goodreads_reviews.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop(['date_updated', 'read_at', 'started_at'], axis=1)
df.info()
```

```python
df.tail(2)
```

```python
# Cambiamos cada user_id por un número subsecuente con enumerate y un list comprehension
user_id_map = {uid: f"{i+1:05d}" for i, uid in enumerate(df['user_id'].unique())}

df['user_id'] = df['user_id'].map(user_id_map)
```

```python
df.head()
```

```python
# Función para generar usuarios
def generar_user():
    letras = ''.join(random.choices(string.ascii_lowercase, k=6))
    numeros = f"{random.randint(0, 999):03d}"
    return f"{letras}_{numeros}"

df['user'] = [generar_user() for _ in range(len(df))]
df.head(2)
```

```python
cols = list(df.columns)
cols.insert(1, cols.pop(cols.index('user')))
df = df[cols]
df.head(2)
```

```python
review_id_map = {rid: f"{i+1:05d}" for i, rid in enumerate(df['review_id'].unique())}

df['review_id'] = df['review_id'].map(review_id_map)
df.head(2)
```

```python
df_books = pd.read_csv('../data/Goodreads_books_with_genres.csv')
df_books.head(2)
```

```python
book_ids_unicos = df['book_id'].unique()
ids_validos = df_books['Book Id'].unique()
cantidad_validos = len(ids_validos)
book_ids_permitidos = book_ids_unicos[:cantidad_validos]

# Filtrar df para conservar solo las filas con book_id en los permitidos
df_filtrado = df[df['book_id'].isin(book_ids_permitidos)].copy()

# Crear el mapa de reemplazo y aplicar
book_id_map = {original: nuevo for original, nuevo in zip(book_ids_permitidos, ids_validos)}
df_filtrado['book_id'] = df_filtrado['book_id'].map(book_id_map)
df = df_filtrado
df.head(2)
```

```python
#No pude convertir la fecha a formato datetime y cambié mejor los valores
n = len(df)
fechas_random = pd.to_datetime(
    np.random.randint(
        pd.Timestamp('2011-01-01').value // 10**9,
        pd.Timestamp('2025-03-31').value // 10**9,
        size=n
    ),
    unit='s'
)

df['date_added'] = pd.to_datetime(fechas_random).strftime('%d/%m/%Y')
df.head(4)
```
